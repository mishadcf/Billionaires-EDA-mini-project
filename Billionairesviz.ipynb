```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
hist=pd.read_csv('/Users/user/Downloads/histbillionaires.csv')
```

```python
age_ranges = hist['age_range_label']
frequencies = hist['count_of_billionaires']



plt.bar(age_ranges, frequencies, edgecolor='black')


plt.xlabel('Age Range')
plt.ylabel('Frequency')
plt.title("Histogram of Billionaires' Age Distribution")


plt.xticks(rotation=90)


plt.show()
```
